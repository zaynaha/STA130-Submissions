The Simple Linear Regression model is expressed as:


$$\Large Y_i = \beta_0 + \beta_1 x_i + \epsilon_i \quad \text{ where } \quad \epsilon_i \sim \mathcal N\left(0, \sigma\right)$$


where:

- **$( Y_i $)** is the *outcome variable* (the value weâ€™re predicting) for the $(i$)-th observation.
- **$( x_i $)** is the *predictor variable* for the $(i$)-th observation, which helps explain changes in $( Y_i $).
- **$( \beta_0 $)** is the *intercept coefficient*, representing the expected value of $( Y $) when  $( x_i = 0 $).
- **$( \beta_1 $)** is the *slope coefficient*, which describes the expected change in $( Y_i $) for a one-unit increase in $( x_i $).
- **$( \epsilon_i $)** is the *error term*, accounting for the difference between the predicted and actual values of $( Y_i $) due to random variability.

In this model, we assume that the error term $( \epsilon_i $) follows a normal distribution:

$[
\epsilon_i \sim \mathcal{N}(0, \sigma)
$]

This means that $( \epsilon_i $) has a mean of 0 and a standard deviation $( \sigma $), which implies that each $( Y_i $) value is normally distributed around the predicted line $( \beta_0 + \beta_1 x_i $) with some random variability.

**Demonstation Below:** 
This code demonstrates how $( Y $) is normally distributed around the line $( \beta_0 + \beta_1 x $) due to the random variation introduced by $( \epsilon $). The plot shows the observed data points alongside the true regression line.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set random seed for reproducibility
np.random.seed(42)

# Define parameters for the Simple Linear Regression model
beta_0 = 5    # Intercept coefficient
beta_1 = 2    # Slope coefficient
sigma = 1     # Standard deviation of the error term

# Generate predictor variable (x) values
n = 100  # Number of data points
x = np.linspace(0, 10, n)

# Generate error term (epsilon) from a normal distribution
epsilon = norm.rvs(0, sigma, size=n)

# Calculate outcome variable (Y) using the regression equation
y = beta_0 + beta_1 * x + epsilon

# Plot the generated data points
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color="blue", label="Observed data (Y)")
plt.plot(x, beta_0 + beta_1 * x, color="red", label="True regression line (Y = Î²â‚€ + Î²â‚X)")
plt.xlabel("Predictor Variable (X)")
plt.ylabel("Outcome Variable (Y)")
plt.title("Simple Linear Regression Model: Y = Î²â‚€ + Î²â‚X + Îµ")
plt.legend()
plt.show()
````

This line represents the true relationship between the predictor `x` and the outcome `Y` if we knew the exact population parameters \( \beta_0 \) and \( \beta_1 \) without any random variation. 2. **Fitted Regression Line**: This is the line generated by fitting a linear model to the simulated sample data. ### Explanation

In the context of Simple Linear Regression, we are interested in how the fitted model predicts values of $Å¶$ (the predicted values) based on the predictor variable $x$.

   $$
   Y_i = Î²_0 + Î²_1 * x_i + Îµ_i \quad \text{where} \quad Îµ_i \sim N(0, Ïƒ)

2. **Fitted Model**: After fitting our model to the sample data, we obtain estimates of the coefficients, denoted as $\hat{\beta}_0$ and $\hat{\beta}_1$. The fitted regression equation becomes:
   $$
   \hat{y}_i = \hat{\beta}_0 + \hat{\beta}_1 x_i
   $$

   \hat{y}_i = \hat{\beta}_0 + \hat{\beta}_1 x_i
   $$

### Fitted Model Using Ordinary Least Squares

The Ordinary Least Squares (OLS) method chooses the line of best fit for a regression model by minimizing the sum of the squared differences between the observed values and the predicted values, known as residuals. In the context of a simple linear regression, these residuals are defined as:

$$
e_i = \hat{\epsilon}_i = Y_i - \hat{Y}_i
$$

Where:
- $(Y_i$) is the observed value.
- $(\hat{Y}_i$) is the predicted value from the fitted model.

The reason it involves "squares" is that squaring the residuals (i.e., $((Y_i - \hat{Y}_i)^2$)) eliminates negative values and emphasizes larger discrepancies. This approach ensures that larger errors have a disproportionately higher influence on the model fitting process, thus leading to a more accurate representation of the data.

The fitted line is derived from this minimization process, which results in estimates for the slope and intercept of the regression line. ### Interpreting Model Accuracy and Proportion of Variation Explained in Simple Linear Regression

The following expressions are connected by their ability to measure the "proportion of variation in $( Y $)" explained by the model. Each provides insight into how well the modelâ€™s predictions (i.e., `fitted_model.fittedvalues`) approximate the actual values of the outcome variable $( Y $).

1. **Expression:** 
   $$[
   1 - \frac{\sum (Y - \text{fitted_model.fittedvalues})^2}{\sum (Y - Y.\text{mean()})^2}
   $$]
   This expression calculates the **proportion of variance in $( Y $) explained by the model**. Hereâ€™s how it works:
   - The numerator, $(\sum (Y - \text{fitted_model.fittedvalues})^2$), represents the **sum of squared residuals (SSR)**, or the variation in $( Y $) not explained by the model.
   - The denominator, $(\sum (Y - Y.\text{mean()})^2$), represents the **total variation in $( Y $)**, or the sum of squared deviations of $( Y $) from its mean.
   - By subtracting the ratio of SSR to total variation from 1, we get the fraction of total variation in $( Y $) that is explained by the model. This proportion is often referred to as $( R^2 $), which serves as a measure of the modelâ€™s explanatory power.

2. **`fitted_model.rsquared`:** 
   This is a direct calculation of $( R^2 $) in the statsmodels library and measures **the accuracy of the model** in explaining variation in $( Y $). It aligns with the above expression by quantifying the proportion of variance in $( Y $) explained by the model. Higher values of $( R^2 $) (close to 1) indicate that the modelâ€™s predictions closely match the actual data.

3. **`np.corrcoef(Y, fitted_model.fittedvalues)[0,1]**2`:**
   This expression computes the square of the **correlation coefficient** between $( Y $) and the modelâ€™s fitted values. Since the correlation measures the strength and direction of a linear relationship, squaring it provides a measure of the **proportion of shared variance** between $( Y $) and the fitted values. For a simple linear regression, this squared correlation coefficient is equivalent to $( R^2 $), capturing the accuracy of the model.

4. **`np.corrcoef(Y, x)[0,1]**2`:**
   This square of the correlation coefficient between $( Y $) and $( x $) represents the **proportion of variation in $( Y $) explained by the linear relationship with $( x $)**. In simple linear regression, this value will also equal $( R^2 $), as it reflects the direct linear association between the predictor variable $( x $) and the outcome $( Y $).

Together, these expressions demonstrate how much of the variation in $( Y $) is accounted for by the predictor variable $( x $) and the fitted model. **Two key assumptions of the Simple Linear Regression model that don't align well with the example data are linearity and homoscedasticity.

**Linearity:** The Simple Linear Regression model assumes that the relationship between the predictor variable (amount of fertilizer) and the outcome variable (crop yield) is linear. However, this example data suggests a non-linear relationshipâ€”as the amount of fertilizer increases, crop yield rises sharply, especially at higher fertilizer levels. This pattern indicates that a straight line may not capture the underlying trend effectively, as seen from the data's shape in the scatter plot.

**Homoscedasticity (Constant Variance of Errors):** The model assumes that residuals (errors) have constant variance across all levels of the predictor variable. In this case, however, the variability in crop yield increases as the amount of fertilizer rises. This increase in residual variance (heteroscedasticity) indicates that errors are not consistent across values of the predictor, which violates the homoscedasticity assumption.

These violations suggest that a different model, such as a polynomial regression or another type of non-linear regression, might be more suitable for capturing the true relationship in this data.

###  What would a model look like in accordance to the assumptions?
In a model where homoscedasticity (constant variance of errors) holds true, the residuals (errors) should ideally have the same level of spread across all values of the predictor (amount of fertilizer, in this case). This means that as fertilizer use increases, the variability in crop yield should not increase but instead remain roughly consistent around the fitted regression line.

**How This Would Look in a Graph:**

Scatter Plot: If the assumption of homoscedasticity is met, the spread of data points around the regression line should be uniform across all values of the predictor variable. The points should be randomly scattered around the line without a fanning pattern or increasing spread as we move along the x-axis.

Residuals Plot: The histogram of residuals should ideally look normally distributed with a peak at zero and tapering off symmetrically. Additionally, if we plot residuals versus the predictor variable, we would expect to see no pattern (i.e., just a "cloud" of points around the horizontal axis), indicating that the residual variance remains constant.

**Generating Homoscedastic Data** 

To make the data align with these assumptions, we can create a dataset where:
- The relationship between fertilizer and yield is linear.
- The residuals (errors) have a constant, normally distributed variance regardless of fertilizer amount.

Let's say:

We keep the slope (beta1) around 40 (for simplicity) and the intercept (beta0) at 20.
We add normally distributed noise to each observation with a constant standard deviation, say, 30, to ensure the spread of errors stays the same across all values of x.

**Here's an example of this code and data generation in Python:**

````python
import numpy as np
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots
from scipy import stats

# Define constants
n = 50  # Number of observations
beta0, beta1 = 20, 40  # Intercept and slope
sigma = 30  # Constant standard deviation for homoscedasticity

# Generate fertilizer amounts (predictor variable)
x = np.linspace(1, 10, n)

# Generate crop yields with homoscedastic residuals (constant variance)
errors = np.random.normal(0, sigma, n)
Y = beta0 + beta1 * x + errors

# Create DataFrame
df = pd.DataFrame({'Amount of Fertilizer (kg)': x, 'Crop Yield (tons)': Y})

# Create scatter plot with trendline
fig1 = px.scatter(df, x='Amount of Fertilizer (kg)', y='Crop Yield (tons)', trendline='ols', title='Crop Yield vs. Amount of Fertilizer (Homoscedastic)')

# Perform linear regression and calculate residuals
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Amount of Fertilizer (kg)'], df['Crop Yield (tons)'])
y_hat = intercept + slope * df['Amount of Fertilizer (kg)']
df['Residuals'] = df['Crop Yield (tons)'] - y_hat

# Plot residuals
fig2 = px.histogram(df, x='Residuals', nbins=10, title='Histogram of Residuals (Homoscedastic)', labels={'Residuals': 'Residuals'})

# Combine scatter plot and residual histogram
fig = make_subplots(rows=1, cols=2, subplot_titles=('Crop Yield vs. Amount of Fertilizer', 'Histogram of Residuals'))
for trace in fig1.data:
    fig.add_trace(trace, row=1, col=1)
for trace in fig2.data:
    fig.add_trace(trace, row=1, col=2)
fig.update_layout(title='Scatter Plot and Histogram of Residuals (Homoscedastic)', xaxis_title='Amount of Fertilizer (kg)', yaxis_title='Crop Yield (tons)', xaxis2_title='Residuals', yaxis2_title='Frequency', showlegend=False)

fig.show(renderer="png")
````

To test for a "no linear association" hypothesis between the waiting time and the eruption duration in the Old Faithful Geyser dataset, weâ€™ll specify our null hypothesis and examine the results of our model fitting.

### Null Hypothesis
The null hypothesis ($(H_0$)) is that there is no linear association, on average, between the waiting time before an eruption (`waiting`) and the duration of the eruption (`duration`). In terms of the Simple Linear Regression model, this hypothesis can be written as:
$$[
H_0: \beta_1 = 0
$$]
where $(\beta_1$) represents the slope coefficient for `waiting` in predicting `duration`. If $(\beta_1 = 0$), it indicates no average linear association between the two variables.

### Running the Model and Interpreting Results
With the code provided, we fit a linear regression model to estimate the relationship between `waiting` and `duration`. Hereâ€™s a breakdown of the interpretation:

1. **p-value of the Slope Coefficient**: The p-value associated with $(\beta_1$) (the `waiting` variable) tests the likelihood of observing a relationship as strong as the one in the data if, in reality, there was no linear association (i.e., if $(H_0$) is true).

2. **Evidence Strength Based on p-value**:
   - **If $( p > 0.1 $)**: No evidence against $(H_0$), suggesting we fail to reject the null hypothesis, indicating no significant linear relationship.
   - **If $( 0.1 \geq p > 0.05 $)**: Weak evidence against $(H_0$).
   - **If $( 0.05 \geq p > 0.01 $)**: Moderate evidence against $(H_0$).
   - **If $( 0.01 \geq p > 0.001 $)**: Strong evidence against $(H_0$).
   - **If $( p \leq 0.001 $)**: Very strong evidence against $(H_0$), suggesting a significant linear association.

````python
import seaborn as sns
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

# Step 2: Fit the original linear regression model to get the observed slope
linear_for_specification = 'duration ~ waiting'
model = smf.ols(linear_for_specification, data=old_faithful)
fitted_model = model.fit()
observed_slope = fitted_model.params['waiting']
print(f"Observed Slope (from Old Faithful data): {observed_slope}")

# Step 3: Bootstrap under the null hypothesis (slope = 0)
n_bootstrap_samples = 1000
bootstrapped_slopes = []
np.random.seed(42)  # For reproducibility

for _ in range(n_bootstrap_samples):
    # Generate data with a null slope (beta1 = 0)
    intercept = np.mean(old_faithful['duration'])
    noise = np.random.normal(0, fitted_model.resid.std(), size=len(old_faithful))
    simulated_duration = intercept + noise

    # Fit model to the simulated data
    simulated_data = pd.DataFrame({'waiting': old_faithful['waiting'], 'duration': simulated_duration})
    bootstrap_model = smf.ols(linear_for_specification, data=simulated_data).fit()
    bootstrapped_slopes.append(bootstrap_model.params['waiting'])

# Step 4: Calculate the p-value for the observed slope within the bootstrapped null distribution
bootstrapped_slopes = np.array(bootstrapped_slopes)
p_value = np.mean(np.abs(bootstrapped_slopes) >= np.abs(observed_slope))  # Two-sided test
print(f"Bootstrapped P-Value: {p_value}")

# Step 5: Plot the bootstrap distribution with the observed slope
plt.hist(bootstrapped_slopes, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(observed_slope, color='red', linestyle='--', linewidth=2, label=f'Observed Slope = {observed_slope:.4f}')
plt.title('Bootstrapped Null Distribution of Slopes')
plt.xlabel('Slope Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Step 6: Interpret based on p-value
if p_value < 0.05:
    print("Reject the null hypothesis: There is significant evidence of a relationship between waiting time and duration.")
else:
    print("Fail to reject the null hypothesis: There is no significant relationship between waiting time and duration.")
````

By rejecting the null hypothesis, we conclude that there is evidence of a relationship between waiting time and duration in the Old Faithful Geyser data. In other words, the slope is statistically significantly different from zero, which implies that waiting time and duration are indeed correlated.

### Model Specification with Indicator Variable
For this model, we use an **indicator variable** to split wait times into two categories, â€œshortâ€ (< 68) and â€œlongâ€ (â‰¥ 68). In this specification, the response variable $( Y_i $) (eruption duration) depends on:

- An intercept term $( \beta_{\text{intercept}} $), representing the average duration for "short" wait times.
- An indicator variable for "long" wait times, $( 1[\text{"long"}](k_i) $), which applies a coefficient $( \beta_{\text{contrast}} $) to "long" wait times.
- The error term $( \epsilon_i $), which is assumed to follow a normal distribution with mean 0 and variance $( \sigma $).

So, the model becomes:
$[
Y_i = \beta_{\text{intercept}} + 1[\text{"long"}](k_i) \beta_{\text{contrast}} + \epsilon_i
$]

This model allows us to capture a possible difference in eruption durations for short versus long wait times without assuming a continuous linear relationship with waiting time.

---

### Key Differences from Previous Models
1. **Continuous vs. Categorical Predictor**:
   - The previous models (1, 2, and 3) used `waiting` as a continuous predictor. This approach assumed a continuous linear relationship between waiting time and eruption duration across all data, or within the short or long wait groups.
   - This new model, however, treats waiting time as categorical, with only two levels ("short" and "long"). This specification assumes that there may be a meaningful difference in eruption durations between these two wait time categories but does not model any within-group linear trend.

2. **Data Grouping**:
   - **Model 1** used all data points with a continuous predictor.
   - **Models 2 and 3** split the data into two separate regressions for short and long wait times but still assumed a continuous relationship within each group.
   - The new model uses a single regression, with a categorical variable to indicate â€œshortâ€ or â€œlongâ€ groups, thus not assuming any within-group continuous trend.

3. **Interpretation of Coefficients**:
   - In the previous models, the slope coefficient represented the rate of change in eruption duration per unit increase in wait time.
   - In this new model, the coefficient $( \beta_{\text{contrast}} $) represents the difference in average eruption durations between â€œlongâ€ and â€œshortâ€ wait times, offering a more straightforward comparison of group means.

---

### Testing the Null Hypothesis: "No Difference Between Groups on Average"
Since we are comparing two independent categories (short and long wait times), **bootstrapping** is a suitable method to estimate the confidence interval for the mean difference. Hereâ€™s a step-by-step approach:

1. **Calculate the Observed Mean Difference**: Compute the actual difference in mean eruption durations between the â€œshortâ€ and â€œlongâ€ groups.
2. **Bootstrap Simulation for Null Hypothesis**: Resample the data with replacement under the null hypothesis (which assumes no difference between groups) to generate a distribution of mean differences.
3. **Confidence Interval**: Use the bootstrapped distribution to compute a 95% confidence interval for the mean difference.
4. **Compare the Observed Difference**: Check if the observed mean difference falls outside this confidence interval. If it does, this provides evidence to reject the null hypothesis of no difference between groups.

---

### Visualization Plan
- **Histogram of Bootstrapped Differences**: Overlay the observed difference as a vertical line on the bootstrapped distribution.
- **95% Confidence Interval**: Highlight the 95% confidence interval range on the histogram to visualize whether the observed difference lies within or outside it.

This approach will allow us to visually and statistically assess if thereâ€™s a significant difference in eruption duration between short and long wait times, based on the indicator variable model specification.


````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns

# Load the geyser dataset from seaborn and rename columns for clarity
old_faithful = sns.load_dataset('geyser').rename(columns={'waiting': 'waiting', 'duration': 'duration'})

# Define "short" and "long" wait categories based on the new threshold
old_faithful['kind'] = np.where(old_faithful['waiting'] < 68, 'short', 'long')

# Ensure 'kind' is treated as a categorical variable
old_faithful['kind'] = pd.Categorical(old_faithful['kind'], categories=['short', 'long'])

# Fit the indicator variable model
model = smf.ols('duration ~ C(kind)', data=old_faithful).fit()

# Get the observed mean difference between "long" and "short"
observed_difference = model.params['C(kind)[T.long]']

# Bootstrapping parameters
n_bootstraps = 10000
bootstrapped_diffs = []

# Bootstrapping loop: shuffle and compute the mean difference under the null hypothesis
for _ in range(n_bootstraps):
    shuffled = old_faithful['duration'].sample(frac=1, replace=True).reset_index(drop=True)
    short_mean = shuffled[old_faithful['kind'] == 'short'].mean()
    long_mean = shuffled[old_faithful['kind'] == 'long'].mean()
    bootstrapped_diffs.append(long_mean - short_mean)

# Calculate 95% confidence interval
conf_interval = np.percentile(bootstrapped_diffs, [2.5, 97.5])

# Plotting the bootstrapped differences
plt.figure(figsize=(10, 6))
plt.hist(bootstrapped_diffs, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(observed_difference, color='red', linestyle='--', linewidth=2, label=f'Observed Difference = {observed_difference:.2f}')
plt.axvline(conf_interval[0], color='black', linestyle=':', linewidth=2, label=f'95% CI = [{conf_interval[0]:.2f}, {conf_interval[1]:.2f}]')
plt.axvline(conf_interval[1], color='black', linestyle=':', linewidth=2)
plt.xlabel('Bootstrapped Mean Differences')
plt.ylabel('Density')
plt.title('Bootstrap Distribution of Mean Differences (Null Hypothesis)')
plt.legend()
plt.show()
````

````python
# A
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the geyser dataset from seaborn and define "short" and "long" categories
old_faithful = sns.load_dataset('geyser')
old_faithful['kind'] = np.where(old_faithful['waiting'] < 68, 'short', 'long')

# Compute the observed difference in means between "long" and "short" groups
observed_diff = old_faithful[old_faithful['kind'] == 'long']['duration'].mean() - \
                old_faithful[old_faithful['kind'] == 'short']['duration'].mean()

# Set up permutation testing parameters
n_permutations = 10000
perm_diffs = []

# Permutation test by shuffling labels
for _ in range(n_permutations):
    shuffled_labels = np.random.permutation(old_faithful['kind'])
    shuffled_diff = old_faithful[shuffled_labels == 'long']['duration'].mean() - \
                    old_faithful[shuffled_labels == 'short']['duration'].mean()
    perm_diffs.append(shuffled_diff)

# Calculate the p-value based on the permutation distribution
p_value = np.sum(np.abs(perm_diffs) >= np.abs(observed_diff)) / n_permutations

# Plot the permutation distribution
plt.figure(figsize=(10, 6))
plt.hist(perm_diffs, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(observed_diff, color='red', linestyle='--', linewidth=2, label=f'Observed Difference = {observed_diff:.2f}')
plt.xlabel('Permutation Differences')
plt.ylabel('Density')
plt.title(f'Permutation Test for Mean Difference (p-value = {p_value:.4f})')
plt.legend()
plt.show()

print(f"Observed Difference: {observed_diff:.2f}")
print(f"P-value from Permutation Test: {p_value:.4f}")
````

````python
# B
import numpy as np
import plotly.express as px
import seaborn as sns

# Load the Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

# Define the two groups based on wait times
short_wait = old_faithful[old_faithful['kind'] == 'short']['duration']
long_wait = old_faithful[old_faithful['kind'] == 'long']['duration']

# Set up bootstrap parameters
n_bootstraps = 10000
bootstrapped_mean_diffs = []

# Perform bootstrapping
for _ in range(n_bootstraps):
    short_sample = np.random.choice(short_wait, size=len(short_wait), replace=True)
    long_sample = np.random.choice(long_wait, size=len(long_wait), replace=True)
    mean_diff = long_sample.mean() - short_sample.mean()
    bootstrapped_mean_diffs.append(mean_diff)

# Calculate the 95% confidence interval
conf_interval = np.quantile(bootstrapped_mean_diffs, [0.025, 0.975])

# Calculate the observed mean difference
observed_mean_diff = long_wait.mean() - short_wait.mean()

# Visualization of bootstrapped mean differences
fig = px.histogram(
    x=bootstrapped_mean_diffs,
    title='Distribution of Bootstrapped Mean Differences',
    labels={'x': 'Bootstrapped Mean Difference'},
    nbins=50
)

# Add lines for the confidence interval
fig.add_vline(x=conf_interval[0], line_color='red', line_dash='dash', 
              annotation_text='Lower CI', annotation_position='top left')
fig.add_vline(x=conf_interval[1], line_color='blue', line_dash='dash', 
              annotation_text='Upper CI', annotation_position='top right')

# Add line for the observed mean difference
fig.add_vline(x=observed_mean_diff, line_color='green', line_dash='solid', 
              annotation_text='Observed Mean Difference', annotation_position='top right')

# Show the plot
fig.show(renderer="png")  # Use this for GitHub and MarkUs submissions
````

The two simulationsâ€”permutation testing and bootstrappingâ€”are distinct approaches to evaluate hypotheses and confidence intervals about group differences. Hereâ€™s how each one works in the context of comparing the mean durations between "short" and "long" wait times:

### (A) Permutation Test

**Goal**: Test the null hypothesis \( H_0: \mu_{\text{short}} = \mu_{\text{long}} \), which states there is no difference in mean duration between the "short" and "long" wait groups.

**Method**:
1. **Shuffle Labels**: The assumption here is that if there is truly no difference in durations between "short" and "long" groups, then the labels "short" and "long" are interchangeable.
2. **Reassign Group Labels**: Combine the duration values from both groups, then randomly shuffle (or permute) these values and split them back into two groups of the original sizes.
3. **Calculate Differences**: Compute the mean difference between the new â€œshortâ€ and â€œlongâ€ groups for each permutation.
4. **Build a Distribution of Differences**: Repeat the shuffling process many times (e.g., 10,000 iterations) to create a distribution of mean differences under the null hypothesis.
5. **Compare Observed Statistic**: The actual observed difference between the "short" and "long" groups is then compared to this permutation distribution. The p-value is calculated as the proportion of permuted differences that are as extreme as or more extreme than the observed difference.

**Interpretation**: If the observed mean difference is extreme within this distribution, it suggests that the observed difference is unlikely under the null hypothesis, providing evidence to reject \( H_0 \).

---

### (B) Bootstrap Confidence Interval

**Goal**: Estimate a 95% confidence interval for the difference in mean durations between "short" and "long" wait times.

**Method**:
1. **Resample within Each Group**: Treat the "short" and "long" groups as separate populations. For each bootstrap iteration, sample with replacement from the "short" group and then separately from the "long" group, ensuring the resampled data is the same size as the original group.
2. **Calculate the Difference**: Compute the difference in mean durations between the bootstrapped "short" and "long" samples.
3. **Build Distribution of Mean Differences**: Repeat this process (e.g., 10,000 times) to create a distribution of mean differences between the bootstrapped samples.
4. **Determine Confidence Interval**: Use the 2.5th and 97.5th percentiles of the bootstrapped mean differences as the bounds for the 95% confidence interval.

**Interpretation**: If the 95% confidence interval for the difference in means excludes 0, it indicates that the difference between "short" and "long" wait times is statistically significant. This approach is powerful because it doesnâ€™t assume a specific distribution for the data, relying instead on the empirical distribution created from resampling.

---

### Comparison of Approaches

- **Permutation Test**: Assesses how likely the observed group difference is under the null hypothesis by simulating the null distribution. Itâ€™s used primarily for hypothesis testing and determining significance.
- **Bootstrap Confidence Interval**: Estimates the range of plausible values for the difference in means by resampling each group. This provides a confidence interval that quantifies the uncertainty around the observed difference.

Together, these simulations help us test for evidence against the null hypothesis and quantify our confidence in the observed group difference without relying on strict normality assumptions.


### A. Permutation Test
**Method Overview:**
- The permutation test evaluates the null hypothesis $( H_0: \mu_{short} = \mu_{long} $) by randomly shuffling the labels (duration values) between the two groups.
- By comparing the observed mean difference to the distribution of mean differences generated from these permutations, we can assess whether the observed difference is statistically significant.

**Key Characteristics:**
- **Non-parametric**: The permutation test does not assume a specific distribution for the data (e.g., normality), making it applicable to various situations.
- **Focus on Label Shuffling**: It directly tests whether the observed differences could arise under the null hypothesis by comparing observed and permuted data.
- **P-value Calculation**: The significance is determined by calculating the proportion of permuted mean differences that are as extreme or more extreme than the observed mean difference.

### B. Bootstrap Confidence Interval
**Method Overview:**
- The bootstrap method involves repeatedly sampling from the observed data (with replacement) within each group to create a distribution of mean differences.
- This method calculates a 95% confidence interval for the difference in means, providing a range of plausible values for the true mean difference.

**Key Characteristics:**
- **Resampling Technique**: Bootstrapping focuses on resampling from the observed data to estimate the distribution of a statistic (mean difference in this case).
- **Confidence Intervals**: The primary output is a confidence interval, providing an estimate of the uncertainty associated with the difference in means.
- **Applicability**: Like the permutation test, it does not assume normality and can be used with smaller sample sizes.
