### Intended Purpose of Setting Up a Null Hypothesis
The primary purpose of establishing a null hypothesis (H) is to create a foundational assumption that can be examined in contrast to an alternative hypothesis (H or H). Essentially, the null hypothesis embodies the existing state of affairs, which researchers strive to contest. By formulating a null hypothesis, researchers can employ statistical tests to evaluate if theres sufficient evidence to reject it, thereby bolstering the alternative hypothesis and enabling data-driven decision-making.

### Key Factor for Examining Ideas Statistically
The distinction between ideas that can be statistically tested and those that cannot hinges on their measurability. An idea is subject to statistical analysis if it can be quantified and evaluated using data. Conversely, non-measurable ideas are not amenable to statistical scrutiny.

### Criteria for a Good Null Hypothesis
A robust null hypothesis (H) should reflect the status quo or default position. It typically makes a claim asserting there is no effect or no difference, and it must be testable through statistical methods. A clearly defined null hypothesis offers a solid framework for comparison against the alternative hypothesis.

### Difference Between Null and Alternative Hypothesis
In hypothesis testing, the null hypothesis (H) conveys the presumption that no significant effect or difference exists within the study's context. In contrast, the alternative hypothesis (H or H) indicates the potential for an effect or difference. Essentially, the null hypothesis is what researchers aim to test and potentially refute, while the alternative hypothesis points to a new direction or deviation from the null.

### Applying this to the Kissing Couples Study (Gntrkn, 2003)

In this context:
- The observed proportion of couples tilting their heads right is 64.5%, based on the sample size of 124 couples. So, we can think of each couple's head tilt as an $x_i$  where $x_i = 1$ for right tilt, and $x_i = 0$ for left tilt.
- The sample mean, $\bar{x}$, is 64.5% or 0.645. This is our observed proportion of couples tilting right.
- The population mean $\mu$ represents the true proportion of all couples tilting their heads to the right.
- The null hypothesis, $H_0$, suggests that there is no preference for left or right tilt, i.e., $\mu_0 = 0.5$ (a 50/50 coin-flipping model).

### Step 1: "Imagine a world where the null hypothesis is true"

We assume that there is no head tilt preference when kissing, so $H_0: \mu = 0.5$. This means couples are equally likely to tilt their heads to the left or right, and the probability of either tilt is 50%. The sampling distribution for this hypothesis can be modeled as flipping a fair coin for each couple.

### Step 2: Simulate the p-value

- To compute a p-value, we simulate the head tilts of 124 couples under the null hypothesis (50% chance of tilting right).
- We repeat this simulation many times, say 10,000 times, and calculate the proportion of right tilts for each simulation.
- After collecting these simulated proportions, we can check how often a proportion as extreme as 64.5% or more occurs under the null hypothesis.

### Step 3: Interpret the p-value

- If the p-value is small, this means that getting a proportion as extreme as 64.5% under the assumption of a 50/50 head tilt is unlikely. This would provide strong evidence against the null hypothesis.
- Conversely, a larger p-value would suggest that the observed proportion is consistent with random variation, and we wouldn't have strong evidence to reject the null hypothesis.

### Step 4: Strength of Evidence Table

Refer to the given "strength of evidence" table to guide your decision based on the computed p-value. For example:
- If the p-value is less than 0.01, this would indicate **very strong evidence** against $H_0$.
- If the p-value is between 0.01 and 0.05, it would suggest **strong evidence** against $H_0$.
- If the p-value is between 0.05 and 0.1, there would be **moderate evidence** against $H_0$.
- If the p-value is greater than 0.1, there would be **weak or no evidence** against $H_0$.

By following these steps, you'll be able to simulate and calculate a p-value, and use that to make an informed decision on whether to reject or fail to reject the null hypothesis based on the strength of the evidence.

                                                          

```python
import numpy as np
import plotly.figure_factory as ff
import plotly.graph_objects as go

# Simulating the coin flips under null hypothesis (H0: No head tilt preference)
np.random.seed(130)  # make simulation reproducible
number_of_simulations = 1000  # Number of bootstrap samples
n_couples = 124  # The number of couples observed in the kissing study
observed_proportion = 0.645  # 64.5% tilted right in the observed data

# Bootstrap sampling distribution under the null hypothesis (50% tilt right, 50% tilt left)
bootstrap_proportions = np.zeros(number_of_simulations)

for i in range(number_of_simulations):
    # Simulate coin flips for each couple (1 = right tilt, 0 = left tilt)
    simulated_flips = np.random.binomial(n=1, p=0.5, size=n_couples)
    bootstrap_proportions[i] = np.mean(simulated_flips)  # Calculate mean (proportion of right tilts)

# Calculating p-value for observed test statistic
p_value = np.mean(np.abs(bootstrap_proportions - 0.5) >= np.abs(observed_proportion - 0.5))
print(f"P-value: {p_value}")

# Calculating 95% confidence intervals (two-tailed test)
ci_lower = np.percentile(bootstrap_proportions, 2.5)
ci_upper = np.percentile(bootstrap_proportions, 97.5)

# Plot the histogram of the bootstrapped sampling distribution
hist_data = [bootstrap_proportions]
group_labels = ['Bootstrapped<br>Sampling Distribution<br>Under Null Hypothesis']
fig = ff.create_distplot(hist_data, group_labels, show_hist=True, show_rug=False, bin_size=0.01)

# Add a line for the lower confidence interval
ci_y = 10  # Adjust height as needed
fig.add_shape(type="line", x0=ci_lower, y0=0, x1=ci_lower, y1=ci_y,
              line=dict(color="Red", width=2), name="95% CI Lower")
# Add a line for the upper confidence interval
fig.add_shape(type="line", x0=ci_upper, y0=0, x1=ci_upper, y1=ci_y,
              line=dict(color="Red", width=2), name="95% CI Upper")
# Add the observed test statistic on the plot
fig.add_shape(type="line", x0=observed_proportion, y0=0, x1=observed_proportion, y1=ci_y,
              line=dict(color="Green", width=3, dash="dash"), name="Observed Test Statistic")
# Add a transparent rectangle for the confidence interval region
fig.add_shape(type="rect", x0=ci_lower, y0=0, x1=ci_upper, y1=ci_y,
              fillcolor="LightSkyBlue", opacity=0.5, line_width=0)
# Add annotations for the confidence interval lines and observed statistic
fig.add_trace(go.Scatter(x=[ci_lower, ci_upper, observed_proportion], 
              y=[ci_y+1, ci_y+1, ci_y+1],  
              text=["95% CI Lower", "95% CI Upper", "Observed Proportion"], 
              mode="text", showlegend=False))

# Update layout with titles
fig.update_layout(
    title="Bootstrapped Sampling Distribution Under Null Hypothesis (50/50) with Observed Statistic",
    xaxis_title="Proportion of Couples Tilting Right", yaxis_title="Density"
)

fig.show(renderer="png") 
```

### Estimating the P-value:
- The **p-value** measures the probability of observing a test statistic as extreme as, or more extreme than, the one actually observed, assuming the null hypothesis is true.
- If the null hypothesis is $( H_0: p = 0.5 )$, then the observed value $( \hat{p} = 0.645 )$ is far enough from 0.5 to raise suspicion about the null hypothesis.
- Since 0.65 is significantly higher than 0.5, we'd expect the p-value to be **low** because values as extreme as 0.65 are unlikely under the assumption of a 50/50 distribution.

### How low might the p-value be?
For a rough estimation:
- In a binomial distribution, under $( H_0: p = 0.5 )$, the **sampling distribution** of the proportion for $( n = 124 )$ couples has a standard error given by:
  $[
  SE = \sqrt{\frac{p(1-p)}{n}} = \sqrt{\frac{0.5(1-0.5)}{124}} \approx 0.045
  ]$
- The observed difference \( 0.645 - 0.5 = 0.145 \) is about **3.22 standard errors** away from the null hypothesized proportion:
 $ [
  z = \frac{0.645 - 0.5}{0.045} \approx 3.22
  ]$
- A z-score of 3.22 corresponds to a **two-tailed p-value** close to 0.001 (or 0.1%), which is extremely small.

### Inferences from this:
1. **Low P-value**: A p-value around 0.001 suggests that it is highly unlikely to observe a proportion of 0.645 or more extreme if there truly was no head tilt preference.
2. **Rejection of the Null Hypothesis**: This low p-value provides strong evidence **against** the null hypothesis. It indicates that the observed data (64.5% tilting right) is very unlikely to have occurred by chance under a 50/50 head tilt scenario.
3. **Tilt Preference Exists**: The data strongly suggest that there **is** a head tilt preference in the population. In this case, the observed preference is towards tilting the head to the right during kissing.

### Conclusion:
Based on the observed test statistic (0.645) and the null hypothesis (0.5), the p-value would likely be very low (around 0.001), meaning there is strong evidence against the null hypothesis. Therefore, we can infer that humans may have a significant tendency to tilt their heads to the right when kissing, rather than it being an equal chance of left or right tilt.

### Understanding P-values in Hypothesis Testing

A p-value is a statistical measure that helps us evaluate the strength of evidence against the null hypothesis. However, it does not provide definitive proof regarding the truth of a hypothesis.

1. **Smaller P-value and Null Hypothesis**:
   - A smaller p-value indicates stronger evidence against the null hypothesis. It suggests that the observed data is less likely under the assumption that the null hypothesis is true.
   - However, a small p-value does not definitively prove that the null hypothesis is false. It only implies that there is a significant level of evidence against it.

2. **Proving Fido's Innocence**:
   - In the context of Fido, the null hypothesis states that he is innocent. A low p-value could suggest that the evidence against him (e.g., the presence of a bin on his head) is strong.
   - Nonetheless, we cannot definitively prove Fido's innocence just because the p-value is high. A high p-value indicates insufficient evidence against the null hypothesis, but it does not confirm his innocence.

3. **Proving Fido's Guilt**:
   - Similarly, we cannot definitively prove Fido's guilt using a low p-value alone. A low p-value indicates that the data would be very unlikely if Fido were innocent, but it does not establish guilt beyond a reasonable doubt.

4. **P-value Thresholds**:
   - There is no specific p-value threshold that can definitively prove either innocence or guilt. Commonly used significance levels (e.g., \(0.05\) or \(0.01\)) help to determine if the evidence is strong enough to reject the null hypothesis, but they do not provide absolute certainty.
   - Conclusions drawn from p-values should always consider the context of the hypothesis, the study design, and other evidence available.

### Conclusion
In summary, p-values and confidence intervals provide valuable insights into the likelihood of observing the data under the null hypothesis but cannot definitively prove or disprove a hypothesis. In Fido's case, both innocence and guilt remain possibilities that cannot be confirmed solely through these statistical measures.

## Key Changes in the Code

### Removal of the Symmetric Test Statistic:
- In a two-tailed test, we consider both extremes of the distribution (both lower and upper tails). Therefore, we calculated a symmetric test statistic and included it in the visualization.
- For a one-tailed test, we only focus on one side (either upper or lower). Thus, the symmetric test statistic and related visualization elements are removed.

### Adjustment of p-value Visualization:
- In a one-tailed test, the p-value represents the probability of observing a test statistic as extreme as (or more extreme than) the observed value, but only in one direction (either greater than or less than, depending on the hypothesis).

### Transparent Rectangle:
- The rectangle indicating the lower extreme region is removed as it is not relevant for a one-tailed test, where we focus solely on the upper extreme region.

## Interpretation Changes

### Hypothesis Testing:
- In a one-tailed test, the null hypothesis is tested against an alternative hypothesis that specifies the direction of the effect (e.g., "greater than" or "less than").
- This approach allows for a more focused test and can lead to a more decisive conclusion regarding the hypothesis in the specified direction.

### Expectation of p-value Comparison

#### p-value Size:
- Generally, we expect the p-value for a one-tailed test to be smaller than that of a two-tailed test if the alternative hypothesis is true. This is because the one-tailed test only considers one tail of the distribution, concentrating the area that contributes to the p-value. In contrast, a two-tailed test splits the alpha level across both tails, resulting in a larger p-value.

## Conclusion
By focusing the hypothesis test in one direction, the one-tailed test can provide more powerful conclusions regarding the significance of results in the specified direction. Its crucial to choose the type of test based on the research question before conducting the analysis.

### Problem Introduction
The famous story of Ronald Fisher and Dr. Muriel Bristol highlights a curiosity about how the order of adding milk and tea affects flavor perception. In this adapted version, we conduct a statistical analysis using a sample of 80 STA130 students to determine whether the order of pouring milk or tea significantly affects the students' ability to identify which was poured first. Specifically, we aim to analyze whether the observed performance (49 correct responses out of 80) is statistically significant or could be attributed to chance.
## Relationship Between This Experiment and the Original with Fisher and Bristol
In the original experiment by Fisher, Dr. Bristol accurately identified the order of pouring milk and tea in a blind taste test, prompting Fisher to question whether her success was due to chance or a real difference in taste. In our adaptation, we seek to analyze a larger group of STA130 students, examining their ability to discern the pouring order in a similar blind taste test. The differences lie in the sample size and the nature of the populationwhile Bristol's preferences were personal, our analysis involves a more generalized assessment of taste perception among a student population.
### Statements of the Null Hypothesis and Alternative Hypothesis

**Null Hypothesis**  
$H_0$: The proportion of STA130 students correctly identifying the order of pouring milk and tea is equal to the proportion expected by chance (i.e., 50%). Mathematically, we express this as:

$$
H_0: p = 0.5
$$

**Informal Interpretation of $H_0$**: There is no difference in the students' ability to tell whether the milk or tea was poured first; their performance could be attributed to random guessing.

**Alternative Hypothesis**  
$H_A$: The proportion of STA130 students correctly identifying the order is greater than what would be expected by chance. This can be expressed as:

$$
H_A: p > 0.5
$$
## Quantitative Analysis
To evaluate the null hypothesis, we will perform a hypothesis test using a one-sample proportion test. We will simulate a sampling distribution under the null hypothesis to estimate the p-value based on the observed test statistic.
#### Methodology Code and Explanations
Here's how we will conduct the analysis using Python with relevant code comments:

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
n_students = 80  # total number of students
n_correct = 49   # number of correct responses

# Sample proportion
p_hat = n_correct / n_students

# Null hypothesis proportion
p_null = 0.5

# Calculate the standard error under the null hypothesis
standard_error = np.sqrt((p_null * (1 - p_null)) / n_students)

# Calculate the z-test statistic
z_statistic = (p_hat - p_null) / standard_error

# Calculate the p-value for the one-tailed test
p_value = 1 - stats.norm.cdf(z_statistic)

# Print the results
print(f"Sample Proportion (p_hat): {p_hat:.2f}")
print(f"Z-Statistic: {z_statistic:.2f}")
print(f"P-Value: {p_value:.4f}")

# Visualize the results
# Generate the null distribution for visualization
null_distribution = np.random.normal(loc=p_null, scale=standard_error, size=10000)

plt.figure(figsize=(10, 6))
plt.hist(null_distribution, bins=30, alpha=0.5, label='Null Distribution', color='lightblue')
plt.axvline(p_hat, color='red', linestyle='dashed', linewidth=2, label='Sample Proportion')
plt.title("Null Distribution of Sample Proportion Under $H_0$")
plt.xlabel("Proportion")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

### Supporting Visualizations
The histogram above illustrates the null distribution of the sample proportion under the null hypothesis $H_0$ (where the true proportion is 0.5). The red dashed line represents the observed sample proportion (49 out of 80).

In simple terms, while more than half of the students correctly identified the pouring order, this result could plausibly be due to chance.

### Ethical Professionalism Considerations
    
Using p-values and hypothesis testing appropriately is an important ethical and professional responsibility of anyone doing data analysis. Actually, there is quite the quiet Contra-Versy (or is it Con-TROV-ersy?) around p-values. First, on a general level, it seems quite clear that p-values and hypothesis testing methodologies MUST play some ongoing contributing role in the so-called "replication crisis" rampantly afflicting mordern science; namely, "significant findings" made in scientific studies are not able to be reproduced by future studies at an alarming rate; and, this whole paradigm of "significant findings" is based on p-values and hypothesis testing... so, something's going on with this methodology in some way...
    
More specifically however, p-values are themselves quite problematic. To see this, just briefly consider the following article titles...

- [Why are p-values controversial?](https://www.tandfonline.com/doi/full/10.1080/00031305.2016.1277161) 
- [What a nerdy debate about p-values shows about science and how to fix it](https://www.vox.com/science-and-health/2017/7/31/16021654/p-values-statistical-significance-redefine-0005)
- [The reign of the p-value is over: what alternative analyses could we employ to fill the power vacuum?](https://royalsocietypublishing.org/doi/10.1098/rsbl.2019.0174)
- [Scientists rise up against statistical significance](https://www.nature.com/articles/d41586-019-00857-9)
- [Statistics experts urge scientists to rethink the p-value](https://www.spectrumnews.org/news/statistics-experts-urge-scientists-rethink-p-value)

While the issues here are relatively advanced and subtle (as introduced [here](https://www2.stat.duke.edu/~berger/p-values.html), presented [here](https://www.jarad.me/courses/stat587Eng/slides/Inference/I06-Pvalues/why_pvalues_dont_mean_what_you_think_they_mean.pdf), and demonstrated using simulation [here](https://jaradniemi.shinyapps.io/pvalue/)), the problem essentially comes down to the fact that most scientists (or just people) don't know how to really interpret the numeric value of a p-value. There are therefore two current proposed solutions to address this challenge.
    
1. Just interpreting p-values using the follwing table (which really isn't that hard, so it's surprising that this solution isn't more broadly adopted...)
    
|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|
    

2. Only do **hypothesis testing** on the basis of confidence intervals, not **p-values** (which might be the best solution wherever doing so is a realistic, convenient  possibility...)

    2. We do not prove $H_0$ is true, we instead do not have evidence to reject $H_0$: "We fail to reject the null hypothesis with a p-value of XYZ"<br><br>
