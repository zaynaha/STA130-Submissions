# 1(a) Classification Decision Tree

A **Classification Decision Tree** is a type of machine learning model used to solve **classification problems**, where the goal is to predict a **categorical (discrete)** outcome based on input features. It works by splitting the data into branches based on feature values, ultimately assigning a class label to each data point at the leaf nodes.

## Problem Addressed
Classification Decision Trees are best suited for problems where you need to:
1. **Categorize or classify data into groups.**
   - Example: Deciding whether an email is spam or not (binary classification).
2. **Handle both numerical and categorical data.**
   - Example: Predicting the type of flower based on petal length and width.
3. **Provide clear and interpretable decision-making logic.**
   - Example: Determining loan approval based on specific financial criteria.

### Key Characteristics
- They address **supervised learning** tasks.
- The outcome variable (target) is **categorical.**
- They produce **if-then rules** that are easy to understand.

---

## Classification vs. Regression

- **Classification**: The target variable is **categorical** (e.g., yes/no, spam/not spam, disease/no disease).
  - Example: Predicting whether a customer will buy a product (yes/no).
  
- **Regression**: The target variable is **continuous** (e.g., predicting a price, temperature, or height).
  - Example: Predicting the price of a house based on features like area, number of bedrooms, etc.

### In Summary:
- **Classification** is used when the outcome is a **category** (discrete values).
- **Regression** is used when the outcome is a **continuous value**.

---

## Real-World Applications

1. **Healthcare:**
   - Diagnosing diseases based on symptoms or test results.
     - Example: Classifying a patient as diabetic or non-diabetic based on glucose levels, BMI, etc.
   - Predicting hospital readmission risks.

2. **Finance:**
   - Approving or rejecting loan applications based on applicant features like income, credit score, and employment history.
   - Detecting fraudulent transactions (fraud vs. legitimate).

3. **Retail:**
   - Predicting customer churn (e.g., whether a customer is likely to stop using a subscription service).
   - Recommending products by categorizing customers into preference groups.

4. **Education:**
   - Predicting whether a student will pass or fail based on attendance, homework completion, and quiz scores.

5. **Customer Service:**
   - Categorizing customer complaints into predefined categories (e.g., billing issue, technical problem).


# 1(b) How Classification Decision Tree Makes Predictions vs. Multiple Linear Regression

### Classification Decision Tree Prediction Process:
A **Classification Decision Tree** makes predictions by following a **series of splits** based on input features. Here's how it works:

1. **Splitting the Data**: 
   - The tree divides the data into subsets using feature values. Each node of the tree represents a decision based on a feature value (e.g., "Is age > 30?" or "Is income > $50,000?").
   - The tree uses a **greedy approach** to choose splits that best separate the data into different classes (using criteria like Gini Impurity or Information Gain).

2. **Following the Branches**:
   - Once the tree is built, to make a prediction for a new data point, the model starts at the **root node** and travels down the branches, answering the decision rules at each node (e.g., "Is income > $50,000?").
   - It keeps moving through the tree until it reaches a **leaf node**, which represents the predicted class label.

3. **Assigning a Class**:
   - The class prediction is based on the majority class of the training data in that leaf node.
   - For example, if most of the data points in the leaf node are labeled "spam," the model will predict "spam" for the new data point.

### Multiple Linear Regression Prediction Process:
**Multiple Linear Regression** makes predictions using a **linear equation** that combines multiple features into a single continuous value. Here's how it works:

1. **Fitting the Line**:
   - Multiple Linear Regression fits a **linear equation** of the form:
     $[
     y = b_0 + b_1x_1 + b_2x_2 + ... + b_nx_n
     $]
     where:
     - $(y$) is the predicted outcome,
     - $(b_0$) is the intercept,
     - $(b_1, b_2, ..., b_n$) are the coefficients (weights),
     - $(x_1, x_2, ..., x_n$) are the input features.
   - The goal of the model is to find the best-fitting line (or hyperplane in higher dimensions) by minimizing the difference between the predicted values and actual values (using a method like **Ordinary Least Squares**).

2. **Making a Prediction**:
   - Once the model is trained, to make a prediction for a new data point, the model simply calculates the value of $(y$) using the learned coefficients and the features of the new data point.
   - This results in a **continuous output** (e.g., a price, temperature, or any other real number).

### Key Differences:
- **Classification Decision Tree**: 
   - Makes predictions by navigating through a tree structure and selecting the most common class in the leaf node (output is categorical).
   
- **Multiple Linear Regression**: 
   - Makes predictions by calculating a weighted sum of the features and applying a linear equation (output is continuous).

In summary, **Classification Decision Trees** predict based on hierarchical decision-making and class majority in leaf nodes, whereas **Multiple Linear Regression** predicts using a weighted sum of feature values within a linear model.

# 2. Real-World Application Scenarios for Classification Metrics

### 1. **Accuracy**

- **Definition**: Accuracy measures how often the model is correct, including both true positives (TP) and true negatives (TN). It’s the proportion of correct predictions out of all predictions made.
  
- **Formula**: 
  $[
  \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
  $]

- **Real-World Application**: 
  **General classification problems** where the class distribution is fairly balanced and the cost of false positives and false negatives is not significantly different.
  
  **Example**: 
  - **Email spam detection**: If the system is reasonably balanced between spam and non-spam emails, accuracy can give a good overview of the model’s performance in distinguishing between the two.
  
- **Rationale**: Accuracy is ideal in cases where the classes are balanced and each correct prediction (whether positive or negative) is equally important. However, it might be misleading if the data is imbalanced (e.g., predicting rare diseases).

---

### 2. **Sensitivity (Recall)**

- **Definition**: Sensitivity, or recall, measures the proportion of actual positives (true instances of the target class) that are correctly identified by the model.
  
- **Formula**:
  $[
  \text{Sensitivity} = \frac{TP}{TP + FN}
  $]

- **Real-World Application**: 
  **Medical diagnostics** where it's crucial to identify as many actual positives as possible, even if it means occasionally classifying negatives as positives.
  
  **Example**: 
  - **Cancer detection**: If a test is identifying whether someone has cancer, we want high sensitivity to ensure we detect as many cancer cases as possible, even if it means some non-cancer cases are misclassified as cancer.
  
- **Rationale**: Sensitivity is important when missing a positive case (e.g., not detecting a disease) could have severe consequences. It prioritizes identifying all potential positive cases, even at the cost of occasional false positives.

---

### 3. **Specificity**

- **Definition**: Specificity measures how well the model identifies actual negatives, or how often the model correctly classifies negative instances.
  
- **Formula**:
  $[
  \text{Specificity} = \frac{TN}{TN + FP}
  $]

- **Real-World Application**:
  **Quality control** or **fraud detection** where it’s important to ensure that negative cases (e.g., non-fraudulent transactions or non-defective products) are identified correctly.

  **Example**: 
  - **Fraud detection in banking**: For a system designed to detect fraudulent transactions, specificity ensures that the system correctly identifies legitimate transactions as non-fraudulent. You don’t want to mistakenly flag too many legitimate transactions as fraud.
  
- **Rationale**: Specificity is useful when you want to ensure the model correctly identifies negatives, which could be important in applications where false positives (e.g., wrongly flagging a non-fraudulent transaction) are costly or inconvenient.

---

### 4. **Precision**

- **Definition**: Precision measures how many of the predicted positives were actually correct. It answers the question: "Of all instances the model predicted as positive, how many were truly positive?"
  
- **Formula**:
  $[
  \text{Precision} = \frac{TP}{TP + FP}
  $]

- **Real-World Application**:
  **Search engines** or **recommendation systems** where you want to ensure that the results returned are highly relevant, even if it means fewer results overall.
  
  **Example**: 
  - **Search engine results**: If a search engine recommends pages or products, you want the recommendations to be highly relevant. High precision means that the pages/products suggested are likely to be what the user is looking for, reducing the chance of irrelevant results.

- **Rationale**: Precision is important when you need to minimize false positives. For instance, in medical testing, a test for a rare disease should only return a positive result when it’s fairly certain the person has the disease. False positives in such cases can lead to unnecessary treatments or anxiety.

---

### Summary Table:

| **Metric**     | **Formula**                                      | **Best Use Case**                                                     | **Why?**                                                                                      |
|----------------|--------------------------------------------------|----------------------------------------------------------------------|----------------------------------------------------------------------------------------------|
| **Accuracy**   | $$(\frac{TP + TN}{TP + TN + FP + FN}$$)            | Balanced classification problems (e.g., spam detection)               | Gives a general idea of correctness when classes are balanced and costs of errors are equal. |
| **Sensitivity**| $(\frac{TP}{TP + FN}$)                           | Medical diagnostics (e.g., cancer detection)                         | Prioritizes detecting positive cases, even at the cost of false positives.                    |
| **Specificity**| $(\frac{TN}{TN + FP}$)                           | Fraud detection (e.g., credit card fraud)                             | Prioritizes correctly identifying negatives, reducing false positives.                       |
| **Precision**  | $(\frac{TP}{TP + FP}$)                           | Search engines (e.g., ensuring relevant results)                      | Minimizes false positives to ensure high-quality results.                                    |

Each metric has its own strengths and weaknesses, and the choice of which one to use depends on the nature of the problem and the costs associated with different types of errors.

### Task Summary:

In this task, we will evaluate the performance of two decision tree models (`clf` and `clf2`) by creating confusion matrices and calculating key metrics such as sensitivity, specificity, and accuracy. These metrics will help assess the models' ability to predict whether a book is hardcover or paperback based on the given features.

### Key Metrics:
- **Sensitivity (Recall)**: The proportion of true positives correctly identified by the model (True Positives / (True Positives + False Negatives)).
- **Specificity**: The proportion of true negatives correctly identified by the model (True Negatives / (True Negatives + False Positives)).
- **Accuracy**: The proportion of correctly classified instances (True Positives + True Negatives) / Total instances.

We will use the `ab_reduced_noNaN_test` dataset, which contains the test data, to evaluate both models.

### Steps:

1. **Make Predictions**:
   - Use `clf.predict()` to make predictions on the `ab_reduced_noNaN_test` dataset.
   - Similarly, use `clf2.predict()` to make predictions on the same dataset.

2. **Generate Confusion Matrices**:
   - Use `confusion_matrix()` from `sklearn.metrics` to generate confusion matrices for both models.

3. **Calculate Metrics**:
   - Calculate sensitivity, specificity, and accuracy for each model using the confusion matrix values.

4. **Visualize the Results**:
   - Visualize the confusion matrices and print the calculated metrics. The values should be rounded to three decimal places for consistency.

````python
# Necessary imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Ensure 'ab_reduced_noNaN_test' is properly structured with 'NumPages', 'Thick', 'List Price', and 'Hard_or_Paper'

# Define the target variable (y_test) and features for the test set
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']
X_test = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]  # Features for clf2
X_test_clf = ab_reduced_noNaN_test[['List Price']]  # Features for clf (only List Price)

# Make predictions using the trained models
y_pred_clf = clf.predict(X_test_clf)  # Only use 'List Price' for clf
y_pred_clf2 = clf2.predict(X_test)   # Use 'NumPages', 'Thick', and 'List Price' for clf2

# Generate confusion matrices for both models
cm_clf = confusion_matrix(y_test, y_pred_clf)
cm_clf2 = confusion_matrix(y_test, y_pred_clf2)

# Calculate sensitivity, specificity, and accuracy for clf
TP_clf = cm_clf[1, 1]  # True Positives for clf
TN_clf = cm_clf[0, 0]  # True Negatives for clf
FP_clf = cm_clf[0, 1]  # False Positives for clf
FN_clf = cm_clf[1, 0]  # False Negatives for clf

sensitivity_clf = TP_clf / (TP_clf + FN_clf)
specificity_clf = TN_clf / (TN_clf + FP_clf)
accuracy_clf = (TP_clf + TN_clf) / (TP_clf + TN_clf + FP_clf + FN_clf)

# Calculate sensitivity, specificity, and accuracy for clf2
TP_clf2 = cm_clf2[1, 1]  # True Positives for clf2
TN_clf2 = cm_clf2[0, 0]  # True Negatives for clf2
FP_clf2 = cm_clf2[0, 1]  # False Positives for clf2
FN_clf2 = cm_clf2[1, 0]  # False Negatives for clf2

sensitivity_clf2 = TP_clf2 / (TP_clf2 + FN_clf2)
specificity_clf2 = TN_clf2 / (TN_clf2 + FP_clf2)
accuracy_clf2 = (TP_clf2 + TN_clf2) / (TP_clf2 + TN_clf2 + FP_clf2 + FN_clf2)

# Round the results to 3 significant digits
sensitivity_clf = np.round(sensitivity_clf, 3)
specificity_clf = np.round(specificity_clf, 3)
accuracy_clf = np.round(accuracy_clf, 3)

sensitivity_clf2 = np.round(sensitivity_clf2, 3)
specificity_clf2 = np.round(specificity_clf2, 3)
accuracy_clf2 = np.round(accuracy_clf2, 3)

# Print the results for both models
print("Model 1 (clf) Performance Metrics:")
print(f"Sensitivity: {sensitivity_clf}")
print(f"Specificity: {specificity_clf}")
print(f"Accuracy: {accuracy_clf}")

print("\nModel 2 (clf2) Performance Metrics:")
print(f"Sensitivity: {sensitivity_clf2}")
print(f"Specificity: {specificity_clf2}")
print(f"Accuracy: {accuracy_clf2}")

# Visualization of confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Display confusion matrix for clf
disp_clf = ConfusionMatrixDisplay(confusion_matrix=cm_clf, display_labels=["Paperback", "Hardcover"])
disp_clf.plot(ax=axes[0], cmap="Blues", values_format=".3f")
axes[0].set_title(f"Confusion Matrix for clf\nAccuracy: {accuracy_clf}")

# Display confusion matrix for clf2
disp_clf2 = ConfusionMatrixDisplay(confusion_matrix=cm_clf2, display_labels=["Paperback", "Hardcover"])
disp_clf2.plot(ax=axes[1], cmap="Blues", values_format=".3f")
axes[1].set_title(f"Confusion Matrix for clf2\nAccuracy: {accuracy_clf2}")

# Show the plots
plt.tight_layout()
plt.show()
````

### Understanding Positive, Negative, and Confusion Matrix in sklearn

#### 1. **What is a "Positive" and "Negative"?**
   - **Positive** refers to the class that is of primary interest in a classification task, typically the "target" class.
     - Example: In binary classification (e.g., predicting whether a book is hardcover or paperback), **Positive** might refer to the **hardcover** class.
   - **Negative** refers to the other class that is not of primary interest.
     - Example: In the same scenario, **Negative** would refer to the **paperback** class.

#### 2. **How to Read an sklearn Confusion Matrix:**
   - A **confusion matrix** is a table used to evaluate the performance of a classification model.
   - It shows the relationship between the actual target values (`y_true`) and the predicted values (`y_pred`).
   
   The matrix is usually organized as follows:

   |                 | Predicted Positive | Predicted Negative |
   |-----------------|--------------------|--------------------|
   | **Actual Positive** | True Positive (TP)  | False Negative (FN)  |
   | **Actual Negative** | False Positive (FP) | True Negative (TN)   |

   - **True Positive (TP)**: The number of correct predictions where the actual class is **positive** and the model predicted **positive**.
   - **True Negative (TN)**: The number of correct predictions where the actual class is **negative** and the model predicted **negative**.
   - **False Positive (FP)**: The number of incorrect predictions where the actual class is **negative** but the model predicted **positive**.
   - **False Negative (FN)**: The number of incorrect predictions where the actual class is **positive** but the model predicted **negative**.

#### 3. **What Leads to TP, TN, FP, and FN:**
   - **TP (True Positive)**: The actual class is positive, and the model correctly predicted positive.
   - **TN (True Negative)**: The actual class is negative, and the model correctly predicted negative.
   - **FP (False Positive)**: The actual class is negative, but the model incorrectly predicted positive (also known as a Type I error).
   - **FN (False Negative)**: The actual class is positive, but the model incorrectly predicted negative (also known as a Type II error).

#### 4. **Which Goes First in the `confusion_matrix` Function?**
   - The correct order for the `confusion_matrix` function in sklearn is:
     ```python
     confusion_matrix(y_true, y_pred)
     ```
     - **`y_true`** refers to the **actual target values**.
     - **`y_pred`** refers to the **predicted values** from your model.

   The confusion matrix is constructed by comparing these values:
   - The rows represent the **actual values** (`y_true`), and
   - The columns represent the **predicted values** (`y_pred`).

To visualize feature importances for a scikit-learn classification decision tree, you can use the feature_importances_ attribute available after fitting the model. This attribute provides the relative importance of each feature in determining the decision tree’s splits. Here's an example of how to visualize it for clf2:

````python
import matplotlib.pyplot as plt
import numpy as np

# Extract feature importances from the fitted clf2 model
feature_importances = clf2.feature_importances_
features = X_train.columns  # Names of the features used in the model

# Create a bar plot for feature importances
plt.figure(figsize=(8, 5))
plt.barh(features, feature_importances, color="skyblue")
plt.xlabel("Feature Importance")
plt.ylabel("Features")
plt.title("Feature Importances for clf2")
plt.xlim(0, np.max(feature_importances) + 0.1)  # Adjust x-axis for better visualization
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
````

### Differences in Interpreting Coefficients in Linear Regression vs. Feature Importances in Decision Trees

In linear regression, coefficients directly indicate how much the predicted outcome changes for a one-unit increase in a predictor variable, assuming all other variables are held constant. This provides a clear and straightforward interpretation of each variable's influence on the model. 

In contrast, feature importances in decision trees measure the relative contribution of each predictor variable to reducing the model's overall prediction error, based on the splits made at decision nodes. Feature importances are not as directly interpretable as coefficients because they do not specify the exact magnitude or direction of a variable's effect, but rather its overall importance to the model's decision-making process.
