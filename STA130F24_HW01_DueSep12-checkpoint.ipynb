````python
Step 1: Understanding df.describe()
What is df.describe()?

df.describe() is a command in Pandas (a Python library) that provides important summary statistics for numeric columns in a dataset.
It gives these values:
count: Number of non-missing entries.
mean: The average value.
std: The standard deviation (how spread out the values are).
min: The smallest value.
25%, 50%, 75%: These are percentiles. For example, 50% is the median, meaning half of the data is below this value.
max: The largest value.
Why are these only for numeric data?

You can’t calculate averages, minimum, or maximum for words or non-numeric data, which is why df.describe() works only with numbers.
Missing data in df.describe()

df.describe() ignores missing data and calculates the statistics based on the available numbers. So if a column has missing values, it just skips them, but still shows the summary stats.
````

````python
Step 2: Handling Missing Data with df.dropna() and del df['col']
What do these commands do?
df.dropna() removes rows where there are missing values.
del df['col'] removes a whole column from your data.
When to use df.dropna()?
Use df.dropna() when there are just a few missing values in rows and you still want to use the remaining columns. It’s good if you want to keep as much data as possible.
When to use del df['col']?
Use del df['col'] if a column has a lot of missing data and isn’t useful anymore. It’s better to remove the whole column than to deal with incomplete information.
Why use del df['col'] before df.dropna()?
If a column is mostly empty and unimportant, deleting it first can save you time and make the df.dropna() process more efficient. It means you only work with useful data.
````
